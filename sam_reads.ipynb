```python
import os
from os import path



if(path.exists("Genus_group_name_WORKING.txt")):
    f_names = open("Genus_group_name_WORKING.txt",'r')
    names_arr= f_names.readlines()
    names_arr= sorted(names_arr)
    f_names.close()
    
else:
    print("Path does not exist.")
    

aln_data =[0]*5000    
    
sam_summary =[]
sam_summary_2=[]
c=0

not_aln=[]
for line in names_arr:
    
    if c ==1:
        break
        
    if line =="":
        break
           
    
    #ide="SAM/"+line[:-1]+".fasta_SAM.sam"
    ide="SAM/Gordonibacter_reads.fasta_SAM.sam"
    if path.exists(ide) ==False:
        print("Path does not exist:"+ PATH_TO_DIR+"/"+line[:-1]+".fasta.aln")
        #break
        continue
    else:
        copen = open(ide,'r')
        sam_data= copen.readlines()
        for data in sam_data:
            
            #amihez illesztett, [2,3,5,12,20]
            split_data =data.split('	')
            # létezik-e a ZS érték?
            if len(split_data) >= 12:
                if split_data[2] == '*':
                    not_aln.append(split_data[0])
                else:
                    if(split_data[12][0]=='Z'):
                        sam_line=(split_data[2],split_data[3],split_data[5],split_data[12].split(':')[-1],split_data[-1].split(':')[-1],len(split_data[9]))
                    else:
                        sam_line=(split_data[2],split_data[3],split_data[5],0,split_data[-1].split(':')[-1],len(split_data[9]))
                sam_summary.append(sam_line)
                
                
                s_e=split_data[0].split('|')[-1].split('S')[1].split('E')
                start= int(s_e[0])
                end=int(s_e[1])
                leng = end-start
                
                if not split_data[0].split('|')[0] == split_data[2]:
                    sam_summary_2.append((leng,int(split_data[3])))
                
                
        copen.close()
            
    c+=1
    
print(len(not_aln))
print(len(sam_summary))
```
