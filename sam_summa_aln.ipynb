```python
import os
from os import path



if(path.exists("Genus_group_name_WORKING.txt")):
    f_names = open("Genus_group_name_WORKING.txt",'r')
    names_arr= f_names.readlines()
    names_arr= sorted(names_arr)
    f_names.close()
    
else:
    print("Path does not exist.")
    
    
summa =[]

c=0
for line in names_arr:
    
    if line =="":
        break
           
    
    ide="summary/"+line[:-1]+"_reads.fasta_summa.txt"
    #ide="summary/Gordonibacter_reads.fasta_summa.txt"
    if path.exists(ide) ==False:
        print("Path does not exist:"+"/"+line[:-1]+"._summa.txt")
        c+=1
        continue
    else:
        s_open= open(ide,'r')
        s_lines=s_open.readlines()
        s_open.close()
        line_counter=0
        hold=0
        hold2=0
        hold3=0
        
        for s in s_lines:
            if s == "":
                break
                
            if line_counter == 2:
                hold = float(s.split('(')[1].split(')')[0][:-1])
            elif line_counter == 3:
                hold2 = float(s.split('(')[1].split(')')[0][:-1])
            elif line_counter == 4:
                hold3 = float(s.split('(')[1].split(')')[0][:-1])
            
            line_counter+=1
            
        triple=(hold, hold2, hold3)
        
        summa.append(triple)
        
        c+=1
print(len(summa))
```

```python
x=[]
y=[]
z=[]
for b in summa:
    
    if b[1]<=10:
        x.append(b[0])
        y.append(b[1])
        z.append(b[2])
   
```

```python

import numpy as np
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [6, 6]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c=z, alpha=1)
ax.set_xlabel('aligned 0 times (%)', fontweight ='bold')
ax.set_ylabel('aligned exactly 1 time (%)', fontweight ='bold')
ax.set_zlabel('aligned >1 times (%)', fontweight ='bold')

print(len(x))
print(len(y))
print(len(z))

ax.view_init(45,45)


plt.show()


```
